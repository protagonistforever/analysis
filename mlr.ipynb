```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn import linear_model
```

```python
area=[2500,3000,3200,3500,4000,4100]
```

```python
bedrooms=[3.0,4.0,None,3.0,5.0,6.0]
```

```python
age=[20,15,18,30,8,8]
```

```python
price=[550000,565000,610000,505000,760000,810000]
```

```python
df=pd.DataFrame({"area":area,"bedrooms":bedrooms,"age":age,"price":price})
```

```python
df
```

```python
import math
```

```python
df.bedrooms.median()
```

```python
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
```

```python

df
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(df.drop("price",axis="columns"),df.price)
```

```python
model.predict([[3000,3,1]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[300,2,1]])
```
